## Sequential API

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential, Model
''
from tensorflow.keras.layers import Flatten, Dense, Input
from tensorflow.keras.optimizers import SGD

import numpy as np
```

### Data Preparation

```python
x_data = np.array( [1, 2, 3, 4, 5, 6] )
t_data = np.array( [3, 4, 5, 6, 7, 8] )

print(x_data.shape, t_data.shape)
```

### Modeling: init W & b randomly

```python
model = Sequential() # 모델

model.add( Flatten(input_shape = (1,)) ) # 입력층
          
model.add( Dense(1, activation = 'linear') ) # 출력층, model.add(Dense(1))        => model.add -> W, b 초기화 되는 부분!!!

# model.add(Dense(1, input_shape = (1,), activation = 'linear')) # 위 코드 한 줄로 압축한 것
```

```python
# 일부러 오류 내려고 만든 것임! 실행 (X)
''' model = Sequential() # 모델

model.add( Flatten(input_shape = (2,)) ) # 입력층
model.add( Dense(2, activation = 'sigmoid') ) '''
```

```python
''' model.summary() '''
# Param  6 => 가중치 4개, 바이어스 2개
```

```python
# w, bias 각각 하나씩 필요한데
print(model.weights)

# numpy=array([[-0.75075364]] => weight
# numpy=array([0.] => bias
```

해당 데이터의 관계는 x_data + 2 = t_data이라
X*w + b 로 생각하면 w=0, b=2로 업데이트 되어야 함

### Compile: Define loss & y

```python
model.compile(optimizer = SGD(), loss = 'mse')

model.summary() # 중간 점검용 디스플레이
```

flatten (Flatten)           (None, 1) => batch자리인데 아직 batch 정해지지 않아서 None  
dense (Dense)               (None, 1)                 2  => (W, b)라서 2개

### Model Learning 

```python
# hist = model.fit(x_data, t_data, epochs = 1000) #
''' Epoch 1/1000
1/1 [==============================]          => batch는 디폴트가 32개인데, 우리 데이터는 6개니까 한번에 들여보낼 수 있으므로 1/1 '''

# hist = model.fit(x_data, t_data, epochs = 1000, batch_size = 3)
''' Epoch 1/1000
2/2 [==============================]          => 데이터 6개니까 batch 3개씩 쪼개면 2번씩 돌아야함 -> 2/2!! 즉, 한 에폭 당 2번씩 돈다는 뜻'''

hist = model.fit(x_data, t_data, epochs = 1000, batch_size = 2)
''' Epoch 1/1000
3/3 [==============================]          => 데이터 6개니까 batch 2개씩 쪼개면 3번씩 돌아야함 -> 2/2!! 즉, 한 에폭 당 3번씩 돈다는 뜻'''

# ** loss 줄어드는지 확인할 것! 
```

```python
print(model.weights)
# W, b가 정답에 가까워짐
```

### Predict

```python
test_input_data = np.array( [-3.1, 3.0, 3.5, 15.0, 20.1] )
label_data = test_input_data + 2.0

result = model.predict(test_input_data)

print(result)
print(label_data.reshape(5,1))

# 정답과 근사한 값을 예측해냄~
```
